```python
import numpy as np
exp = np.exp
import matplotlib.pyplot as plt

Tl = np.linspace(300, 350)

k2b = 3.95e10 * exp(-6864/Tl) # Greer (Jamal)
k2c = 10**(10.99 - 2152/Tl)/1000 # Hikita - The Kinetics of Kinetics of Carbon Dioxide
k2d = exp(12.4 - 904.6/Tl)/1000 # Freguia - Modeling of CO2 Capture by Aqueous Monoethanolamine

plt.plot(Tl, k2b, label='Greer')
plt.plot(Tl, k2c, label='Hikita')
plt.plot(Tl, k2d, label='Freguia')

plt.legend()
```
